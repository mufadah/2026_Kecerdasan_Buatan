#1. Central Tendency

# Mendapatkan Nilai Mean

```python
import numpy as np
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.mean(score)

print(x)
```

#Mendapatkan Nilai Median

```python
import numpy as np
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.median(score)

print(x)
```

#Mendapatkan Nilai Modus

```python
from scipy import stats
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = stats.mode(score, keepdims=True)

print(x.mode[0])
```

#2. Spread

#Mendapatkan Nilai Range

```python
import numpy as np
data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]
minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum

print(x)
```

#Mendapatkan Nilai Precentile

```python
import numpy as np
data = [90, 80, 70, 65, 110, 80, 83, 67, 94]
x = np.percentile(data, 90)

print(x)
```

#Mendapatkan Nilai Quartile dan Interquartile Range

```python
import numpy as np
data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]
Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1
Q1, Q3, IQR

print(Q1)
print(Q3)
print(IQR)
```

#Mendapatkan Nilai Variance dan Standart Deviation

```python
import numpy as np
data= [86, 87, 88, 86, 87, 85, 86]
mean = np.mean(data)
var = np.var(data)
std = np.std(data)
mean, var, std

print(mean)
print(var)
print(std)
```

```python
import numpy as np

nilai_ujian = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99]

# a. Urutkan data dari yang terkecil sampai terbesar
nilai_urut = np.sort(nilai_ujian)
print(f"a. Data terurut:\n{nilai_urut}\n")

# b. Identifikasi Q1, Q3, dan IQR
# Menggunakan np.percentile untuk mendapatkan quartil yang presisi
Q1 = np.percentile(nilai_ujian, 25)
Q3 = np.percentile(nilai_ujian, 75)
IQR = Q3 - Q1
print(f"b. First Quartile (Q1) : {Q1}")
print(f"   Third Quartile (Q3) : {Q3}")
print(f"   Interquartile Range (IQR) : {IQR}\n")

# c. Tentukan nilai upper bound dan lower bound
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
print(f"c. Lower Bound : {lower_bound}")
print(f"   Upper Bound : {upper_bound}\n")

# d. Tentukan data outlier
outliers = [nilai for nilai in nilai_ujian if nilai < lower_bound or nilai > upper_bound]
print(f"d. Nilai ujian yang menjadi outlier: {outliers}")
```

```python
import numpy as np

# Menentukan nilai tengah (midpoint) untuk setiap rentang kelas
# 41-45 -> 43; 46-50 -> 48; 51-55 -> 53; 56-60 -> 58; 61-65 -> 63; 66-70 -> 68
midpoints = np.array([43, 48, 53, 58, 63, 68])

# Frekuensi masing-masing kelas
frekuensi = np.array([2, 3, 6, 4, 5, 2])

# Total data (n)
n = np.sum(frekuensi)

# 1. Hitung Mean (Rata-rata data berkelompok)
mean_kelompok = np.average(midpoints, weights=frekuensi)

# 2. Hitung Variance (menggunakan rumus sampel: dibagi n-1)
# Varians = jumlah dari f * (x - mean)^2 dibagi (n-1)
variance = np.sum(frekuensi * (midpoints - mean_kelompok)**2) / (n - 1)

# 3. Hitung Standar Deviasi (Akar dari variance)
std_dev = np.sqrt(variance)

print("--- Perhitungan Standar Deviasi Data Berkelompok ---")
print(f"Total Frekuensi (n) : {n}")
print(f"Rata-rata (Mean)    : {mean_kelompok:.2f} kg")
print(f"Variance            : {variance:.2f}")
print(f"Standar Deviasi     : {std_dev:.2f} kg")
```
