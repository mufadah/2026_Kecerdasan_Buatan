# 4. Operasi Matriks dan Vektor Dengan Python

#Membuat Matriks Dengan Python

```python
import numpy as np
A = np.array([[80, 30, 25],
[50, 20, 10]])
B = np.array([[30, 60, 15],
[10, 30, 30]])

hasil = A + B
print(hasil)
```

# Melakukan Operasi Dot Product

```python
import numpy as np
A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

hasil = np.dot(A, B)
print(hasil)
```

#Melakukan Operasi Cross Product

```python
import numpy as np
A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

hasil = np.cross(A, B)
print(hasil)
```

```python
import numpy as np

# 1. Definisi Vektor A dan B
# Vektor A = 5i + 3j + 7k -> [5, 3, 7]
# Vektor B = 12i - 3j + 1k -> [12, -3, 1]
A = np.array([5, 3, 7])
B = np.array([12, -3, 1])

# 2. Menghitung Panjang (Magnitude) masing-masing vektor
panjang_A = np.linalg.norm(A)
panjang_B = np.linalg.norm(B)

# 3. Hasil Perkalian Titik (Dot Product)
# Menggunakan np.dot() atau operator @
hasil_dot = np.dot(A, B)

# 4. Hasil Perkalian Silang (Cross Product)
# Menggunakan np.cross()
hasil_cross = np.cross(A, B)

# Menampilkan Hasil
print(f"Panjang Vektor A: {panjang_A:.2f}")
print(f"Panjang Vektor B: {panjang_B:.2f}")
print("-" * 30)
print(f"Hasil Perkalian Titik (A . B): {hasil_dot}")
print(f"Hasil Perkalian Silang (A x B): {hasil_cross}")
```
